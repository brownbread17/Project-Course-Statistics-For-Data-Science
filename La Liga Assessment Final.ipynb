```python
import pandas as pd
import numpy as np
```

```python
liga=pd.read_csv('Laliga.csv')
liga
```

```python
liga=liga.replace('-','0')
liga
```

```python
#to convert the string/object to int type to perform arithemetic operations
liga['Debut']=liga['Debut'].astype(str).str[0:4].astype(int)
#liga_new=liga[(liga['Debut']>1930) & (liga['Debut']<1980)] 
liga_new=liga[liga['Debut'].between(1930,1980)]
liga_new
```

```python
#convert points to float
liga['Points']=liga.Points.astype(float)
#sort in decresing order
liga.sort_values(by='Points',ascending=False)
liga.head(5)
```

```python
def Goal_diff_count():
    
    liga['GoalsFor']=liga.GoalsFor.astype(float)
    liga['GoalsAgainst']=liga.GoalsAgainst.astype(float)
    liga['Goal_Difference']=liga['GoalsFor']-liga['GoalsAgainst']
    df=pd.DataFrame(liga)
    return df
```
